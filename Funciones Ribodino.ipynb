# Solución 1

### Ejercicio 1

Escribir una función para calcular el factorial de un número cualquiera.

```python
x = int(input("Ingrese cualquier número: "))

def factorial (x): 
    for i in range(1,x+1):
        x = x*i

    return(x)
    
factorial(x)
```

### Ejercicio 2

Escribir una función para calcular la suma de una serie comenzando por un número cualquiera y terminando en otro número que debe ser mayor al primero.


```python
x = int(input("Ingrese cualquier número1: "))
j = int(input("Ingrese cualquier número2: "))
a = 0

def sumaNumeros (x,j,a):
    if (j>x):
        for i in range (x,j+1):
            a += i
        print (f'la suma de los numeros que estan entre {x} y {j} es: {a}')
    else:
        return('error')
    
   
sumaNumeros(x,j,a)
```
